```python
#import libraries 

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df=pd.read_csv("movies.csv")
```

```python
#lets look at the data 
df.head()
```

```python
# lets see if there is any missing data 

for col in df.columns:
    pct_missing= np.mean(df[col].isnull())
    print("{} - {}%".format(col,pct_missing))
```

```python
#filling the null values in budget with 0

df['budget']=df['budget'].fillna(0)

df['gross']=df['gross'].fillna(0)


df['votes']=df['votes'].fillna(0)

df['runtime']=df['runtime'].fillna(0)
```

```python
#datatyoes of culumns

df.dtypes
```

```python
#change datatype of columns 

df['budget']=df['budget'].astype('int64')

df['gross']=df['gross'].astype('int64')

df['votes']=df['votes'].astype('int64')

df['runtime']=df['runtime'].astype('int64')
```

```python
df
```

```python
#create correct year column

dateformat=df['released'].astype(str).str.replace(r'\s*\(.*\)', '', regex=True)

```

```python
dateformat
```

```python
dateformat=pd.to_datetime(dateformat, format='%B %d, %Y', errors='coerce')

```

```python
df['correctyear']=(dateformat.dt.year)
```

```python
df.dtypes
```

```python
df['correctyear']=df['correctyear'].fillna(0)
```

```python
df['correctyear']=df['correctyear'].astype('int64')
```

```python
df=df.sort_values(by=['gross'],inplace=False, ascending=False)
```

```python
# drop duplicates 


df.drop_duplicates()
```

```python
#budget has high correlation 
#company has high correlation 
```

```python
# scatter plot with budget vs gross 

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross earnings')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
df.head()
```

```python
# regression plot 

sns.regplot(x='budget', y='gross',data=df, scatter_kws={"color":"black"},line_kws={"color":"blue"})
```

```python
# lets start lookinh at correlation 

df.corr(method='pearson') #pearson, kendall, spearman
```

```python
# high correlation between budget and gross

correlation_matrix=df.corr(method='pearson') 

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation matrix for numeric numbers')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# assigning category codes to the non-numeric data so that it can be easy for the analysis 


df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes

df_numerized
```

```python
# high correlation between budget and gross

correlation_matrix=df_numerized.corr(method='pearson') 

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation matrix for numeric numbers')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```
